```python
import pandas as pd
file_path = 'Student Attitude and Behavior.csv'
student_data = pd.read_csv(file_path)
student_data.head()
```

```python
import sqlite3
conn = sqlite3.connect('Students.db')
cur = conn.cursor()
```

```python
create_table_query = """
CREATE TABLE IF NOT EXISTS StudentInfo (
    CertificationCourse TEXT,
    Gender TEXT,
    Department TEXT,
    HeightCM REAL,
    WeightKG REAL,
    TenthMark REAL,
    TwelfthMark REAL,
    CollegeMark REAL,
    Hobbies TEXT,
    DailyStudyingTime TEXT,
    PreferToStudyIn TEXT,
    SalaryExpectation INTEGER,
    LikeDegree TEXT,
    CareerPursuitBasedOnDegree TEXT,
    SocialMediaVideoTime TEXT,
    TravellingTime TEXT,
    StressLevel TEXT,
    FinancialStatus TEXT,
    PartTimeJob TEXT
);
"""
```

```python
cur.execute(create_table_query)
```

```python
student_data.to_sql('StudentInfo', conn, if_exists='replace', index=False)
```

```python
cur.execute("SELECT * FROM StudentInfo LIMIT 5;")
inserted_rows = cur.fetchall()
```

```python
conn.commit()
conn.close()
```
